# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline
def ReLU(x):
    return abs(x) * (x > 0)

def dReLU(x):
    return (1 * (x > 0))
x = plt.linspace(-10,10,100)
relu=ReLU(x)
drelu=dReLU(x)
```

```python
plt.plot(x, relu, 'r')
plt.show()
```

```python
plt.plot(x, drelu, 'r')
plt.show()
```
